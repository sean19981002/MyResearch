```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import json
```

```python
j = dict()
for i in range(14):
    with open('%d.json' % i, 'r') as fp:
        jj = json.load(fp = fp)
        j.update(jj)

columns = []
for key in j.keys():
    if j[key] != None:
        obj = j[key]
        row = [key, obj['location'], obj['followers_count'], obj['following_count'], obj['tweet_count'], obj['verified'], obj['created_at']]
        columns.append(row)
        #print(row)
col_name = ['id', 'location', 'followers_count', 'following_count', 'tweet_count', 'verified', 'created_at']
df = pd.DataFrame(columns, columns=col_name)
df['class'] = -1
df.loc[df['location'] == 'None', 'class'] = 0
```

```python
usa_db = pd.read_excel('uscities.xlsx', usecols=['city', 'state_id', 'state_name', 'county_name'])
usa_db.head(5), usa_db.tail(5)
```

```python
citys = list(usa_db['city'])
state_id = list(set(usa_db['state_id']))
state_name = list(set(usa_db['state_name']))
usa_list = citys  + state_name
len(usa_list)
```

```python
for index in df.index:
    location = df.loc[index, 'location']
    if 'united state' in location.lower() or 'usa' in location.lower():
        df.loc[index, 'class'] = 1
    for usa_loc in usa_list:
        if usa_loc.lower() in location.lower():
            df.loc[index, 'class'] = 1
    for i in state_id:
        if i in location:
            df.loc[index, 'class'] = 1
    
```
